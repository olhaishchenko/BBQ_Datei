### Aufgabe 2.1: Würfel
Implementiere die Klasse `Wuerfel`.
Sie soll über die Methode `werfen()` verfügen, die ein Zahl zwischen
1 und 6 im Attribut `augen` speichert.

Erzeuge ein Objekt der Klasse `Wuerfel` und würfele hiermit solange, bis der Würfel eine 6 liefert und gib die Anzahl der Versuche aus.

Für den Wurf kann folgendes benutzt werden:
```py

randint(1,6)
```

````python
# Dein Code
from random import randint
# import random as r # r.randint(1,6)
class Wuerfel:
    def __init__(self):
        self.fleache = None

    def werfen(self):
        self.fleache = randint(1, 6)
        return self.fleache
    
wuerf1 = Wuerfel()

i=1
while wuerf1.fleache != 6:
    wuerf1.werfen()
    i += 1
    
print(f"Das war {i} Versuche")

````

### Aufgabe: 2.3 Bruch-Klasse

**Methoden**:

-   `__str__()`: Gibt den Bruch als String im Format "zaehler/nenner" zurück.
-   `kuerzen()`: Kürzt den Bruch auf seine einfachste Form.
-   `erweitern(other: Bruch)`: Passt die Zähler und Nenner des aktuellen Bruchs und des übergebenen Bruchs an, sodass beide Brüche denselben Nenner haben, um die Addition korrekt durchführen zu können. Wenn das übergebene Argument kein `Bruch` ist soll ein `TypeError` ausgelöst werden.

````python
import math

class Bruch:
    def __init__(self, zaehler: int, nenner: int):
        if nenner == 0:
            raise ValueError("Nenner darf nicht 0 sein!")
        self.zaehler = zaehler
        self.nenner = nenner
        self.kuerzen()

    def __str__(self):
        return f"{self.zaehler}/{self.nenner}"

    def __repr__(self):
        return f"Bruch(zaehler={self.zaehler}, nenner={self.nenner})"

    def kuerzen(self):
        """Kürzt den Bruch auf seine einfachste Form"""
        ggT = math.gcd(self.zaehler, self.nenner)
        self.zaehler = self.zaehler // ggT
        self.nenner = self.nenner // ggT
        return self

    def erweitern(self, other):
        """Passt die Zähler und Nenner an, sodass beide Brüche denselben Nenner haben"""
        if not isinstance(other, Bruch):
            raise TypeError("Das Argument muss ein Bruch sein!")
        
        gemeinsamer_nenner = self.nenner * other.nenner
        self.zaehler *= other.nenner
        other.zaehler *= self.nenner
        self.nenner = other.nenner = gemeinsamer_nenner

    def __add__(self, other):
        """Addiert zwei Brüche, nachdem sie auf denselben Nenner erweitert wurden"""
        if not isinstance(other, Bruch):
            raise TypeError("Das Argument muss ein Bruch sein!")

        # Brüche auf denselben Nenner erweitern
        self.erweitern(other)

        # Addiere die Zähler und erstelle einen neuen Bruch
        neuer_zaehler = self.zaehler + other.zaehler
        neuer_nenner = self.nenner
        return Bruch(neuer_zaehler, neuer_nenner)

# Testfälle
bruch1 = Bruch(1, 2)
bruch2 = Bruch(1, 3)

print("Bruch 1:", bruch1)  # Ausgabe: 1/2
print("Bruch 2:", bruch2)  # Ausgabe: 1/3

# Brüche addieren
ergebnis = bruch1 + bruch2
print("Ergebnis der Addition:", ergebnis)  # Ausgabe: 5/6

# Fehlerprüfung (TypError und ValueError)
try:
    bruch3 = Bruch(1, 0)
except ValueError as e:
    print("Fehler:", e)

try:
    ergebnis_fehler = bruch1 + 5  # Sollte ein TypeError auslösen
except TypeError as e:
    print("Fehler:", e)
````
