# Normalverteilung

Die Normal- oder Gauß-Verteilung ist ein wichtiger Typ stetiger Wahrscheinlichkeitsverteilungen. Ihre Wahrscheinlichkeitsdichte wird auch Gauß-Funktion, Gaußsche Normalverteilung, Gaußsche Verteilungskurve, Gauß-Kurve, Gaußsche Glockenkurve, Gaußsche Glockenfunktion, Gauß-Glocke oder schlicht Glockenkurve genannt.

Beispiele mit einer Normalverteilung:

- Zufällige Messfehler,
- Zufällige Abweichungen vom Sollmaß bei der Fertigung von Werkstücken,
- Beschreibung der brownschen Molekularbewegung.
- In der Versicherungsmathematik ist die Normalverteilung geeignet zur Modellierung von Schadensdaten im Bereich mittlerer Schadenshöhen.
- In der Messtechnik wird häufig eine Normalverteilung angesetzt, die die Streuung der Messfehler beschreibt. Hierbei ist von Bedeutung, wie viele Messpunkte innerhalb einer gewissen Streubreite liegen. 

Wir schauen uns zunächst die Definition an, erstellen ein Beispiel mit Python indem wir Scipy und Numpy verwenden. Dabei diskutieren wir die Eigenschaften einer Normalverteilung.

Die Verteilung wird mit der Wahrscheinlichkeitsdichtefunktion (en. **PDF** probability density function) beschrieben:


## $$  f(x,\mu,\sigma) = \frac{1}{\sigma\sqrt{2\pi}}e^\frac{-1}{2z^2} $$

wobei: $$z=\frac{(X-\mu)}{\sigma}$$

mit: 
- μ=Durchschnitt 
- σ=Standardabweichung 
- π=3.14...
- e=2.718... 

Die Gesamtfläche zwischen der Kurve der Wahrscheinlichkeitsdichtefunktion und der X-Achse ist 1. Daher repräsentiert die Fläche unter der Kurve zweier Ordinaten (Senkrechte) X=a und X = b, wobei a<b, die Wahrscheinlichkeit dass X zwischen a und b liegt. Diese Wahrscheinlichkeit kann wie folgt ausgedrückt werden: $$P(a<X<b)$$

----------

Die Normalverteilung hat die folgenden Eigenschaften:

1. Die Kurve ist für die theoretische Verteilung symetrisch  
2. Die Spitze der Kurve ist bei dem Durchschnitt.
3. Die Standardabweichung gibt der Kurve ihre Form:
    -Die Kurve ist für eine höhere Standardabweichung klein und breit.
    -Die Kurve ist für eine niedrigere Standardabweichung schmal und hoch.
4. Die Fläche unter der Kurve ist gleich 1
5. Der Durchschnitt = Median    

In der Praxis sind reale Daten oft nicht perfekt normalverteilt. Reale Verteilungen sind oft nur „annähernd“ normal, da in echten Datensätzen meist kleine Abweichungen auftreten, z. B. leichte Asymmetrien oder zusätzliche „Ausreißer“. Man spricht hier von der „theoretischen Normalverteilung“, um zu betonen, dass diese idealisierte Form oft nur als Modell oder Annäherung für reale Daten dient.

Wenn du die Normalverteilung für eine ganze Population betrachtest, zeigt sie die Verteilung der Datenpunkte über den gesamten Wertebereich der betrachteten Variablen an. Die Normalverteilung beschreibt, wie die Werte einer Variablen innerhalb der Population streuen

Die **3-Sigma-Regel** (auch als **Empirische Regel** bekannt) beschreibt eine wichtige Eigenschaft der Normalverteilung und besagt:

- **Etwa 68%** der Daten liegen innerhalb von **1 Standardabweichung** ($\sigma$) vom Mittelwert ($\mu$).
- **Etwa 95%** der Daten liegen innerhalb von **2 Standardabweichungen** ($2\sigma$) vom Mittelwert.
- **Etwa 99,7%** der Daten liegen innerhalb von **3 Standardabweichungen** ($3\sigma$) vom Mittelwert.

Diese Regel gilt **nur für Normalverteilungen** und ist eine nützliche Faustregel, um schnell einzuschätzen, wie stark Daten um den Mittelwert verteilt sind. Sie hilft, Ausreißer zu erkennen, da Werte außerhalb des Bereichs von $\mu \pm 3\sigma$ in der Regel als ungewöhnlich betrachtet werden.

Die **Formel für die kumulative Verteilungsfunktion (CDF)** der Normalverteilung lautet:

$$F(x)={\frac {1}{\sigma {\sqrt {2\pi }}}}\int _{-\infty }^{x}e^{-{\frac {1}{2}}\left({\frac {t-\mu }{\sigma }}\right)^{2}}\mathrm {d} t,\quad x\in \mathbb {R} $$
<img src="https://i.postimg.cc/8Cfs9pSt/The-Normal-Distribution.png" alt="Bildbeschreibung" width="800" height="800">

Der z-Wert oder z-Score beschreibt den standardisierten Abstand eines Werts vom Mittelwert (Anzahl der Standardabweichungen) und wird wie folgt berechnet:
$$z=\frac{x−μ}{σ}$$
<br>
Der z-Score wird verwendet, wenn die Populationsstandardabweichung bekannt ist. Er gibt an, wie viele Standardabweichungen ein Wert von dem Mittelwert entfernt ist.<br>
Der t-Score wird verwendet, wenn die Populationsstandardabweichung nicht bekannt ist und stattdessen die Stichprobenstandardabweichung verwendet wird.

$$t=\frac{x−μ}{\frac{s}{\sqrt{n}}}$$

- X: Der einzelne Datenpunkt, dessen Score berechnet werden soll.
- μ: Der Mittelwert der gesamten Population.
- σ: Die Standardabweichung der gesamten Population.
- s: Standardabweichung der Stichprobe
- n: Stichprobengröße

### Standardnormalverteilung
Die Standardnormalverteilung ist eine spezielle Form der Normalverteilung mit einem Mittelwert von 0 und einer Standardabweichung von 1. Variablen können standardisiert (normalisiert) werden, um in diese Form überführt zu werden, was den Vergleich zwischen verschiedenen Datensätzen erleichtert (siehe z-Score). 

Erstellen wir uns selbst eine Normalverteilung mit Scipy.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import stats
import numpy as np

mean = 0
std = 1

# Erstelle eine range X
X = np.arange(-4,5,0.1)

# Erstelle eine Normalverteilung für X
Y = stats.norm.pdf(X,mean,std)
plt.ylim([0, .6])
plt.plot(X,Y);
```

Wir visualisieren die normalverteilten Werte mit einem Seaborn Histogram.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 0  # Mittelwert
sigma = 1  # Standardabweichung
x = np.linspace(-4, 4, 1000)  # Werte für die x-Achse

plt.figure(figsize=(12, 8))

# Dichtefunktion der Normalverteilung
y = norm.pdf(x, mu, sigma)
plt.plot(x, y, color='blue')

# Funktion zum Berechnen und Zeichnen des Integrals
def plot_integral(lower_bound, upper_bound, color, label):
    # Berechnung des Integrals
    integral_value = norm.cdf(upper_bound, mu, sigma) - norm.cdf(lower_bound, mu, sigma)
    
    # Füllbereich für das Integral
    x_fill = np.linspace(lower_bound, upper_bound, 1000)
    y_fill = norm.pdf(x_fill, mu, sigma)
    plt.fill_between(x_fill, y_fill, alpha=0.3, color=color, label=f'{label} : {integral_value:.4f}')
    
    # Hinzufügen der Grenzen des Integrals
    plt.axvline(lower_bound, color=color, linestyle='--')
    plt.axvline(upper_bound, color=color, linestyle='--')

# Integrale von -1 bis +1
plot_integral(mu - 1, mu + 1, 'blue', 'Integral von -σ bis +σ')

# Integrale von -2 bis +2
plot_integral(mu - 2, mu + 2, 'green', 'Integral von -2σ bis +2σ')

# Integrale von -3 bis +3
plot_integral(mu - 3, mu + 3, 'red', 'Integral von -3σ bis +3σ')

# Hinzufügen von Beschriftungen und Legende
plt.title('Standard Normalverteilung mit Integralen [-σ, +σ], σ = [1, 2, 3] ')
plt.xlabel('x')
plt.ylabel('Dichte')
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5, linestyle='--')
plt.grid()
plt.legend();
```

### Normalapproximation
Die Normalverteilung kann als Näherung für Binomial- oder andere diskrete Verteilungen verwendet werden, wenn bestimmte Bedingungen erfüllt sind (z. B. große Stichproben oder bestimmte Erfolgswahrscheinlichkeiten). Dies ist hilfreich, um komplexe Berechnungen für diskrete Verteilungen zu vereinfachen

Der Übergang von der Binomialverteilung zur Normalverteilung basiert auf dem zentralen Grenzwertsatz (CLT), der besagt, dass die Binomialverteilung bei einer großen Anzahl von Versuchen n unter bestimmten Bedingungen durch eine Normalverteilung approximiert werden kann. Die Binomialverteilung beschreibt die Anzahl der Erfolge in einer festen Anzahl von unabhängigen Bernoulli-Experimenten mit einer Erfolgswahrscheinlichkeit 𝑝.
<br>
Die Normalverteilung hingegen ist eine kontinuierliche Verteilung, die durch den Mittelwert 
μ und die Standardabweichung 
σ definiert ist. Ihre Dichtefunktion ist eine Glockenkurve, die symmetrisch um den Mittelwert liegt. Die Umwandlung der Binomialverteilung in eine Normalverteilung erfolgt, wenn 
n ausreichend groß ist und 
p einen mittleren Wert hat (d.h., 
p sollte weder sehr klein noch sehr groß sein). Für die Normalverteilung, die die Binomialverteilung approximiert, sind der Mittelwert μ=n⋅p und die Standardabweichung 
σ = n⋅p⋅(1−p).

Die Normalverteilung ist eine gute Annäherung an die Binomialverteilung, wenn bestimmte Bedingungen erfüllt sind, wie zum Beispiel dass n⋅p≥10 und 
n⋅(1−p)≥10, was sicherstellt, dass die Binomialverteilung ausreichend symmetrisch und gut für die Normalverteilungsapproximation geeignet ist. Zudem wird häufig eine Kontinuierlichkeitskorrektur verwendet, da die Binomialverteilung diskret und die Normalverteilung kontinuierlich ist. Dabei wird der Bereich um 
k herum, also von 
k−0.5 bis 
k+0.5, anstelle des einzelnen Werts k verwendet, um die Diskrepanz zwischen den beiden Verteilungen zu verringern.

Beispielsweise, wenn wir ein Experiment mit 
n=100 Versuchen und einer Erfolgswahrscheinlichkeit 
p=0.5 haben, dann ist der Mittelwert 
μ=50 und die Standardabweichung 
σ=5. In diesem Fall kann die Binomialverteilung durch eine Normalverteilung mit diesen Parametern approximiert werden, wenn 
n groß genug ist.

Zusammengefasst lässt sich sagen, dass die Normalverteilung eine gute Approximation für die Binomialverteilung bietet, wenn 
n groß ist und 
p im mittleren Bereich liegt, und dass dabei eine Kontinuierlichkeitskorrektur oft angewendet wird, um die Unterschiede zwischen den diskreten und kontinuierlichen Verteilungen zu verringern.

```python
from scipy.stats import norm, binom

# Parameter
n = 100   # Anzahl der Versuche
p = 0.5   # Erfolgswahrscheinlichkeit
mu = n * p  # Mittelwert
sigma = np.sqrt(n * p * (1 - p))  # Standardabweichung

# Binomialverteilung
x_binom = np.arange(0, n+1)
y_binom = binom.pmf(x_binom, n, p)

# Normalverteilung
x_norm = np.linspace(0, n, 1000)
y_norm = norm.pdf(x_norm, mu, sigma)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x_norm, y_norm, label="Normalverteilung", color="r", lw=2)
plt.stem(x_binom, y_binom, label="Binomialverteilung", basefmt=" ", linefmt="b-", markerfmt="bo") #  erstellt ein Stabdiagramm
plt.title("Binomialverteilung vs. Normalverteilung")
plt.xlabel("Anzahl der Erfolge")
plt.ylabel("Wahrscheinlichkeit")
plt.legend()
plt.show()
```

### Zentrale Grenzwertsatz (ZGS)
Der ZGS ist ein grundlegendes Konzept in der Statistik. Er besagt, dass die Summe oder der Durchschnitt einer großen Anzahl unabhängiger, identisch verteilter Zufallsvariablen sich mit zunehmender Anzahl annähernd einer Normalverteilung (Gauß-Verteilung) nähert – und das, egal wie die ursprüngliche Verteilung dieser Variablen aussieht.

Stell dir vor, du hast eine Gruppe von Zufallsvariablen mit einer beliebigen Verteilung, wie z.B. einer schiefen Verteilung. Wenn du nun immer mehr dieser Variablen nimmst und ihren Durchschnitt bildest, wird die resultierende Verteilung mit einer ausreichend großen Anzahl von Variablen immer näher an eine Normalverteilung kommen, selbst wenn die ursprüngliche Verteilung gar nicht normal ist.

Das bedeutet, dass viele praktische Probleme, die auf zufälligen Prozessen beruhen, durch die Annahme einer Normalverteilung vereinfacht werden können, was die Analyse und Berechnungen erheblich erleichtert.

```python
sample_size = 100  # Größe der Stichprobe
num_samples = 1000  # Anzahl der Stichproben

# Erzeugung einer nicht-normalen Verteilung (Exponentialverteilung)
data = np.random.exponential(scale=1, size=(num_samples, sample_size))
mittelwert = np.mean(data)

# Erstellen der Subplots
plt.figure(figsize=(12, 8))

# Schrittweise Berechnung und Plotten der Mittelwerte
steps = [10, 50, 100, 500, 1000]  # Anzahl der Stichproben zu verschiedenen Zeitpunkten

for i, step in enumerate(steps, 1):
    means = np.mean(data[:step], axis=1)
    
    plt.subplot(3, 2, i)
    sns.histplot(means, bins=30, alpha=0.7, color='blue', edgecolor='black', kde=True)
    plt.title(f"Verteilung der Mittelwerte (nach {step} Stichproben)")
    plt.xlabel("Mittelwert")
    plt.ylabel("Häufigkeit")
    
# Ausgangsverteilung plotten (nach allen Stichproben)
plt.subplot(3, 2, len(steps)+1)
plt.hist(data.flatten(), bins=30, alpha=0.7, color='orange', edgecolor='black')
plt.title(f"Ausgangsverteilung (Exponentialverteilung) Mittelwert={mittelwert}")
plt.xlabel("Werte")
plt.ylabel("Häufigkeit")

plt.tight_layout()
plt.show()
```

### **Z-Standardisierung** (auch Z-Transformation genannt) 
wird durchgeführt, um die beiden Datensätze, die schiefe Verteilungen aufweisen, in eine standardisierte Form zu überführen. 

Die Z-Standardisierung transformiert die Daten so, dass sie eine **Mittelwert von 0** und eine **Standardabweichung von 1** haben. Dies geschieht durch die Berechnung des **Z-Werts** für jeden Wert in den Datensätzen, der die Anzahl der Standardabweichungen angibt, die der Wert vom Mittelwert entfernt ist. Die Formel lautet:

$$
Z = \frac{X - \mu}{\sigma}
$$

wobei $X$ der Originalwert, $\mu$ der Mittelwert und $\sigma$ die Standardabweichung der Daten sind.

Durch die Z-Standardisierung werden Verzerrungen aufgrund unterschiedlicher Skalen oder Einheiten der ursprünglichen Daten beseitigt und die beiden Verteilungen vergleichbar gemacht. Dies ist besonders nützlich, wenn man verschiedene Datensätze oder Variablen in einem Modell vergleichen möchte.

Die beiden KDE-Plots zeigen dann die **ursprünglichen** und die **standardisierten** Verteilungen, wobei die standardisierten Werte nun die gleiche Skala haben.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore


np.random.seed(0)

data1 = np.random.gamma(shape=2.0, scale=1.5, size=500) / .5  # schiefe Verteilung
data2 = np.random.chisquare(df=2, size=500) +10  # ebenfalls schief

# Originale KDE-Plot
plt.figure(figsize=(12, 6))
sns.kdeplot(data1, label="Original Data 1", color="blue", fill=True)
sns.kdeplot(data2, label="Original Data 2", color="red", fill=True)
plt.title("KDE der Originaldaten")
plt.xlabel("Wert")
plt.ylabel("Dichte")
plt.legend()
plt.grid(True)
plt.show()

# Z-Standardisierung (z-score)
data1_z = zscore(data1)
data2_z = zscore(data2)

# KDE-Plot der z-Standardisierten Daten
plt.figure(figsize=(12, 6))
sns.kdeplot(data1_z, label="Z-Standardisiert Data 1", color="blue", linestyle="--")
sns.kdeplot(data2_z, label="Z-Standardisiert Data 2", color="red", linestyle="--")
plt.title("KDE der z-Standardisierten Daten")
plt.xlabel("Z-Wert")
plt.ylabel("Dichte")
plt.legend()
plt.grid(True)
plt.show()
```
