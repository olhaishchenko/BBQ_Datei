### Pareto-Prinzip
Das Pareto-Prinzip, auch bekannt als 80/20-Regel, besagt, dass oft 80 % der Ergebnisse durch 20 % der Ursachen hervorgerufen werden. Diese Regel stammt von dem italienischen Ökonomen Vilfredo Pareto, der feststellte, dass in Italien etwa 20 % der Bevölkerung 80 % des Vermögens besaßen. Seitdem hat sich das Pareto-Prinzip als allgemeines Muster in vielen Bereichen etabliert, von Wirtschaft bis Zeitmanagement.

#### Grundidee des Pareto-Prinzips
Das Pareto-Prinzip beschreibt die Ungleichheit oder Unausgewogenheit in verschiedenen Prozessen und Verteilungen. Es bedeutet nicht immer exakt 80/20, sondern zeigt das Prinzip, dass ein kleiner Teil der Ursachen häufig den größten Teil der Effekte ausmacht. Beispiele hierfür sind:

- Wirtschaft: 20 % der Kunden generieren 80 % des Umsatzes.
- Zeitmanagement: 20 % der Aufgaben bringen 80 % des gewünschten Ergebnisses.
- Qualitätsmanagement: 20 % der Fehlerquellen führen zu 80 % der Probleme.
- Vertrieb: 20 % der Produkte machen 80 % des Gesamtumsatzes aus.

#### Anwendung des Pareto-Prinzips
Das Prinzip wird oft verwendet, um Ressourcen effizienter zu verteilen, indem man sich auf die wichtigsten Faktoren konzentriert. Im Qualitätsmanagement, z. B. bei der Fehlerbehebung, nutzt man das Paretodiagramm. Es hilft, die Ursachen nach Häufigkeit oder Einfluss zu sortieren, um schnell die wichtigsten Bereiche zu identifizieren, an denen man ansetzen sollte.

#### Beispiel eines Paretodiagramms
Ein Paretodiagramm kombiniert ein Balkendiagramm und eine kumulative Kurve:

- Balken: Stellen die Häufigkeit der Ursachen (z. B. Fehlerarten) dar, geordnet von den wichtigsten bis zu den weniger wichtigen.
- Kumulative Kurve: Zeigt den Prozentsatz der Gesamtwirkung an, der durch die Ursachen bis zu diesem Punkt erreicht wird.
Das Ziel besteht darin, die Ursachen zu finden, die den größten Einfluss haben, und sich darauf zu konzentrieren.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Beispielhafte Daten
data = {'Kategorie': ['A', 'B', 'C', 'D', 'E' , 'F' ,'G'],
        'Häufigkeit': [40, 35, 25, 10, 8, 7, 6]}

df = pd.DataFrame(data)

# Daten nach Häufigkeit sortieren und kumulative Prozente berechnen
df = df.sort_values(by='Häufigkeit', ascending=False)
df['Kumulierte Prozent'] = df['Häufigkeit'].cumsum() / df['Häufigkeit'].sum() * 100

# Paretodiagramm erstellen
fig, ax = plt.subplots()

# Balkendiagramm mit Seaborn
sns.barplot(x='Kategorie', y='Häufigkeit', data=df, color='C0', ax=ax)

# Sekundärachse für die kumulative Prozentlinie
ax2 = ax.twinx()    # erzeugt eine zweite y Achse auf der rechten Seite
ax2.plot(df['Kategorie'], df['Kumulierte Prozent'], color='C1', marker='o', ms=7, linestyle='-', label='Kumulierte Prozent')
ax2.axhline(y=80, color='r', linestyle='--', label='80%-Linie')

# Achsen und Labels
ax.set_ylabel('Häufigkeit')
ax2.set_ylabel('Kumulierte Prozent')
ax.set_xlabel('Kategorie')

# Legende außerhalb des Plots rechts anzeigen
fig.legend(loc='center', bbox_to_anchor=(1.02, 0.5), borderaxespad=0)

# Anzeigen des Diagramms
plt.title('Paretodiagramm')
plt.tight_layout(rect=[0, 0, 0.85, 1])  # Platz für die Legende schaffen
plt.show()
```
