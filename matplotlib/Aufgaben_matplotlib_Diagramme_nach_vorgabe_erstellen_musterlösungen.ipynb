```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Aufgabe (1)
x=[1,10,30,49]
# y=[0,30,90,150]
# alternativ
x = np.array(x)
y = 3*x
plt.plot(x,y)
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.axis([0, 50, 0, 160]) # [x-start, x-ende, y-start, y-ende]
plt.tick_params(axis='y', direction='in', right=True)  
plt.tick_params(axis='x', direction='in', top=True)
plt.title("Draw a line")
plt.show()
```

```python
#Aufgabe2
y=[2,4,1]
x=[1,2,3]
plt.plot(x,y)
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.axis([1.0, 3.0, 1.0, 4.0])
plt.xticks([i / 10 for i in range(10,31,5)])
plt.tick_params(axis='y', direction='in', right=True)  
plt.tick_params(axis='x', direction='in', top=True)
plt.tick_params(direction="in")
plt.title('Sample Graph')
plt.show()
```

```python
#Aufgabe3
x1=[10,20,30]
x2=[10,20,30]
y1=[20,40,10]
y2=[40,10,30]
plt.plot(x1, y1, label="line 1")
plt.plot(x2, y2, "g-", label="line 2")
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.axis([10, 30, 10, 40])
plt.xticks([i for i in range(10,31,5)])
legend = plt.legend(loc= "upper right", edgecolor='black')
plt.title("Two or more lines ...")
plt.tick_params(axis='y', direction='in', right=True)  
plt.tick_params(axis='x', direction='in', top=True)
plt.show()
```

```python
#Aufgabe7
t = np.arange(0., 5., 0.2)
# plot() kann mehrere Diagramme auch gleichzeitig plotten
plt.plot(t, t, 'g--', 
         t, t**2, 'bs', 
         t, t**3, 'r^')
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.axis([0, 5, 0, 120])
plt.tick_params(axis='y', direction='in', right=True)  
plt.tick_params(axis='x', direction='in', top=True)
plt.show()
```

```python
# Aufgabe - 5. Seite
liste = [(2, 1, 1), (2, 3, 4), (2, 3, 5), (2, 3, 6)]
for tup in liste:
    plt.subplot(tup[0], tup[1], tup[2])
    plt.xticks(())
    plt.yticks(())
plt.show()
```

Erste Iteration (2, 1, 1):

(2, 1, 1) bedeutet ein Raster mit 2 Zeilen und 1 Spalte, wobei das Plot-Feld 1 belegt wird. Dies erzeugt eine einzelne Spalte mit zwei Zeilen, wobei nur das obere Feld (Position 1) besetzt ist.
Zweite Iteration (2, 3, 4):

(2, 3, 4) bedeutet ein Raster mit 2 Zeilen und 3 Spalten, wobei der Plot in Feld 4 eingefügt wird. Das Raster ist breiter (3 Spalten), und 4 besetzt die erste Position in der zweiten Zeile (erste Spalte unten).
Dritte und vierte Iteration (2, 3, 5) und (2, 3, 6):

Beide ebenfalls in einem 2x3-Raster. Die Werte 5 und 6 legen fest, dass die Plots in den Positionen 5 (zweite Spalte der zweiten Zeile) und 6 (dritte Spalte der zweiten Zeile) platziert werden.
