## Einfache Datenvisualisierung 

#### Übung
Für die ersten 24 Tage eines Monats haben wir den Durchschnitt-Lufttemperaturen in Grad Celsius für alle jeweils drei Tage gemessen und in einer Liste gespeichert:

```python
celsius_values = [25.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
```
Erstelle dafür mit Matplotlib ein Diagramm, welches entlang x-Achse die Tage (z.B. 1 bis 22) und entlang der y-Achse die Temperaturen angezeigt werden.

````python
# schreibe deinen code hier
import matplotlib.pyplot as plt
from math import sin, exp
````

````python
# Daten für die durchschnittlichen Lufttemperaturen und Tage
celsius_values = [25.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
days = list(range(1, 25, 3))  # Tage: 1, 4, 7, ..., 22 (alle drei Tage)

# Diagramm erstellen
plt.figure(figsize=(10, 6))
plt.plot(days, celsius_values, marker='o', color='b', linestyle='-', linewidth=2, markersize=6)

# Diagrammformatierung
plt.title("Durchschnittliche Lufttemperaturen für jeden 3. Tag eines Monats")
plt.xlabel("Tag des Monats")
plt.ylabel("Temperatur (°C)")
plt.xticks(days)
plt.grid(visible=True, linestyle='--', alpha=0.6)

# Diagramm anzeigen
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Задаємо функцію h(t)
def h(t):
    return 10 * np.exp(-5 * (t - 1)**2)

# Перша похідна h'(t)
def h_prime(t):
    return -100 * (t - 1) * np.exp(-5 * (t - 1)**2)

# Друга похідна h''(t)
def h_double_prime(t):
    return (-100 * (1 - 10 * (t - 1)**2)) * np.exp(-5 * (t - 1)**2)

# Задаємо діапазон значень для t
t_values = np.linspace(0, 2, 100)

# Обчислюємо значення функцій для кожного t
h_values = h(t_values)
h_prime_values = h_prime(t_values)
h_double_prime_values = h_double_prime(t_values)

# Створюємо графіки
plt.figure(figsize=(10, 6))

# Графік h(t)
plt.plot(t_values, h_values, label="h(t)", color="blue")
# Графік h'(t)
plt.plot(t_values, h_prime_values, label="h'(t)", color="green")
# Графік h''(t)
plt.plot(t_values, h_double_prime_values, label="h''(t)", color="red")

# Налаштування графіка
plt.title("Графіки функцій h(t), h'(t), h''(t)")
plt.xlabel("t")
plt.ylabel("Значення функцій")
plt.legend()
plt.grid(True)

# Відображення графіка 
plt.show()
````

````python
import numpy as np

# Перша похідна h'(t)
def h_prime(t):
    return -100 * (t - 1) * np.exp(-5 * (t - 1)**2)

# Критичні точки для h'(t)
t_critical_1 = 1 - np.sqrt(1/10)
t_critical_2 = 1 + np.sqrt(1/10)

# Обчислення значень h'(t) в критичних точках
max_speed_1 = h_prime(t_critical_1)
max_speed_2 = h_prime(t_critical_2)

print("Максимальна швидкість у точці t ≈ {:.3f}: {:.3f}".format(t_critical_1, max_speed_1))
print("Максимальна швидкість у точці t ≈ {:.3f}: {:.3f}".format(t_critical_2, max_speed_2))
````
