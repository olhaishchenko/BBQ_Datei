# Funktionen

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def plot_koor(ax):
    ax.axvline(0, alpha=0.2, ls="--")  
    ax.axhline(0, alpha=0.2, ls="--")  
    ax.set_xlim([-10, 10])             
    ax.set_ylim([-10, 10])  
```

### Definition Funktion
Eine Funktion ist eine Beziehung zwischen zwei Variablen, die jedem Wert der unabh√§ngigen Variablen $x$ einen eindeutigen Wert der abh√§ngigen Variablen $y$ zuordnet. <br>
Wenn man in der Beziehung f√ºr einen bestimmten $y$-Wert zwei verschiedene $x$-Werte finden kann, ist es **keine** Funktion.

### Lineare Funktion
Eine lineare Funktion ist eine mathematische Beziehung zwischen zwei Variablen, die durch eine gerade Linie dargestellt werden kann.<br>
$f(x) =m\times x +b$ \
$m$ ist die Steigung (eng. slope) der Gerade, wenn dieser Wert negativ ist, dann bedeutet dies dass die Funktion f√§llt \
$b$ ist der y-Achsenabschnitt (eng. intercept), d.h. der Schnittpunkt der Funktion mit der y- Achse wenn $x = 0$ also $f(0)=b$

```python
f = lambda x, m, b: m * x + b  

x = np.linspace(-10, 10, 100)
y1 = f(x, 2, 5)
y2 = f(x, -3, 10)
y3 = f(x, .5, -1)

fig, ax = plt.subplots( figsize=(8,4))
plot_koor(ax)
ax.plot(x,y1, "b", label = "$f(x) = 2x + 5$")
ax.plot(x,y2, "r", label = "$f(x) = -3x + 10$")
ax.plot(x,y3, "g", label = "$f(x) = 0,5x - 1$")
plt.legend();
```

### Quadratische Funktion
Eine quadratische Funktion ist eine mathematische Beziehung zwischen einer unabh√§ngigen Variablen 
$x$ und einer abh√§ngigen Variablen $y$, die durch ein Polynom zweiten Grades beschrieben wird.<br> 
Die allgemeine Form einer quadratischen Funktion ist 
$f(x)=ax^2+bx+c$, wobei a, b und c Konstanten sind und ùëé ‚â† 0.<br>

Der Graph einer quadratischen Funktion ist eine Parabel, die entweder nach oben (wenn a>0) oder nach unten (wenn 
a<0) ge√∂ffnet ist. Die h√∂chste oder niedrigste Stelle der Parabel wird als Scheitelpunkt bezeichnet und spielt eine wichtige Rolle bei der Bestimmung der Extremwerte der Funktion. Quadratische Funktionen k√∂nnen auch Schnittpunkte mit der x-Achse (Nullstellen) haben, die die L√∂sungen der Gleichung 
$ax^2+bx+c=0$ darstellen.

```python
f = lambda x, a, b, c: a * x**2 + b * x + c  

x = np.linspace(-10, 10, 100)
y1 = f(x, 1, 0,0 )
y2 = f(x, 5, 0 ,-5 )
y3 = f(x, -1, 4, 0)

fig, ax = plt.subplots( figsize=(8,4))
plot_koor(ax)
ax.plot(x,y1, "b", label = "$f(x) = x^2$")
ax.plot(x,y2, "r", label = "$f(x) = 5x^2 - 5$")
ax.plot(x,y3, "g", label = "$f(x) = -x^2 +4x$")
plt.legend();
```

### Polynom
Ein Polynom als Funktion ist eine spezielle Art von mathematischer Funktion, die aus einer endlichen Summe von Potenzen einer Variablen besteht, wobei jede Potenz mit einem Koeffizienten multipliziert wird. Eine allgemeine Polynomfunktion in einer Variablen 
x hat die Form:<br>
$P(x)=a_n * x^n +a_{n‚àí1} *x^{n‚àí1}   +‚Ä¶+a_1 *x+a_0$ 

Hierbei gilt:
- $n$: Der Grad des Polynoms, das die h√∂chste Potenz von $x$ angibt.
- $a_n, a_{n‚àí1},‚Ä¶,a_1,a_0$ : Die Koeffizienten, wobei $a_n$ ‚â† 0

**Eigenschaften**:
- Polynomfunktionen sind √ºber den gesamten Bereich der reellen Zahlen stetig und unendlich oft differenzierbar. Das bedeutet, dass sie keine Spr√ºnge oder L√∂cher im Graphen haben und man die Ableitungen beliebig oft bilden kann.

- Der Graph einer Polynomfunktion ist eine kontinuierliche Kurve. Je nach Grad des Polynoms kann der Graph unterschiedliche Formen annehmen:

    - Polynome ersten Grades (lineare Funktionen) sind gerade Linien.
    - Polynome zweiten Grades (quadratische Funktionen) haben die Form einer Parabel.
    - H√∂here Grade k√∂nnen komplexere Formen annehmen, die mehrere Wendepunkte und Extremstellen aufweisen k√∂nnen.
- Das Verhalten einer Polynomfunktion f√ºr gro√üe Werte von 
x wird durch den h√∂chsten Grad und das Vorzeichen des h√∂chsten Koeffizienten bestimmt:
    - F√ºr gerade Grade hat der Graph an beiden Enden das gleiche Verhalten (beide nach oben oder beide nach unten).
    - F√ºr ungerade Grade hat der Graph an den Enden entgegengesetztes Verhalten (eine Richtung nach oben und die andere nach unten).

```python
f_g3 = lambda x, a, b, c, d: a * x**3 + b * x**2 + c*x + d  
f_g4 = lambda x, a, b, c, d, e: a * x**4 + b * x**3 + c*x**2 + d*x + e  

x = np.linspace(-10, 10, 1000)
y1 = f_g3(x, 2, 3, 0, 4 )
y2 = f_g4(x, -9, -5, 7, -2, 4 )

fig, ax = plt.subplots( figsize=(8,4))
plot_koor(ax)
ax.plot(x,y1, "b", label = "$3.Grad : f(x) = 2*x^3 + 3*x^2 + 4$")
ax.plot(x,y2, "r", label = "$4.Grad : f(x) = -9*x^4 - 5*x^3 + 7x^2 - 2*x + 4$")
plt.xlim([-5,5])
plt.ylim([-10,10])
plt.legend();
```

### Exponentialfunktion
Eine Exponentialfunktion ist eine mathematische Funktion der Form: \
$f(x)=a‚ãÖb^x$ \
  \
Hierbei gilt: \
$a$: Der Anfangswert oder der $y$-Achsenabschnitt. Er gibt den Funktionswert an, wenn $x=0$ ist. \
$b$: Die Basis der Exponentialfunktion, ein positiver Wert (b>0), der nicht 1 sein darf. Er bestimmt die Wachstums- oder Zerfallsrate. \
$x$: Die unabh√§ngige Variable (oft als Zeit oder andere Dimension interpretiert).\
\
Wachstum oder Zerfall: \
Wenn $b>1$, zeigt die Funktion exponentielles Wachstum. Das bedeutet, der Funktionswert nimmt immer schneller zu, je gr√∂√üer $x$ wird. \
Wenn $0<b<1$, zeigt die Funktion exponentiellen Zerfall. Der Funktionswert nimmt immer schneller ab, je gr√∂√üer $x$ wird. <br>
<br>
Exponentialfunktionen sind √ºber den gesamten Bereich der reellen Zahlen stetig und differenzierbar. Das bedeutet, sie haben keine Spr√ºnge oder L√ºcken. \
F√ºr $x‚Üí+‚àû$ (gro√üe positive Werte) strebt die Funktion bei $b>1$ gegen unendlich und bei<br> $0<b<1$ gegen $0$.
F√ºr $x‚Üí‚àí‚àû$ (gro√üe negative Werte) strebt die Funktion bei $b>1$ gegen $0$ und bei $0<b<1$ gegen unendlich.
<br><br>
Exponentialfunktion zur **Basis $e$ (Eulersche Zahl)**:<br> 
Eine besondere Form der Exponentialfunktion ist die zur Basis der Euler‚Äôschen Zahl $e$,<br> 
etwa $f(x)=e^x$<br>
Diese Funktion ist besonders in der Mathematik und Naturwissenschaft verbreitet, da sie eine konstante √Ñnderungsrate besitzt.

```python
f_ex = lambda x, a, b: a * b**x

x = np.linspace(0, 5, 1000)
y1 = f_ex(x,1,2)  
y2 = f_ex(x,1,.5)
y3 = f_ex(x,1,np.e)
y4 = f_ex(-x,1,2)

fig, ax = plt.subplots(ncols=4,figsize=(12,3))

ax[0].set_title("Wachstum")
ax[0].plot(x,y1, label="$f(x) =  2^x$")
ax[0].legend()

ax[1].set_title("Zerfall")
ax[1].plot(x,y2, label="$f(x) =  0,5^x$")
ax[1].legend()

ax[2].set_title("Mit Euler‚Äôschen Zahl")
ax[2].plot(x,y3, label="$f(x) =e^x$")
ax[2].legend();

ax[3].set_title("Mit negativen X-Werten\n(siehe x-Achse)")
ax[3].plot(-x,y4, label="$f(x) =2^{-x}$")
ax[3].legend();
```

### Logarithmusfunktion
Eine Logarithmusfunktion ist eine Funktion, die die Umkehrung der Exponentialfunktion darstellt. Sie hat die allgemeine Form:<br>
$f(x)=log_b(x)$<br>
Dabei gilt:<br>
- $b$: Die Basis des Logarithmus, ein positiver Wert $b>0$, der nicht 1 sein darf.<br>
- $x$: Der Wert, f√ºr den der Logarithmus bestimmt wird, wobei $x>0$ gelten muss.<br>
- $f(x)$: Gibt an, zu welchem Exponenten man die Basis $b$ potenzieren muss, um den Wert $x$ zu erhalten. <br> 
    $f(x)=log_b(x)$ bedeutet, dass $b^{f(x)}=x$ gilt.<br>
<br>

Da der Logarithmus die Umkehrung der Exponentialfunktion ist, gilt:<br>
Wenn $y=b^x$, dann ist $x=log_b(y)$.<br>
Bei einer Basis $b>1$ steigt der Funktionswert $f(x)=log_b(x)$ langsam an.<br>
F√ºr eine Basis $0<b<1$ sinkt $f(x)=log_b(x)$ ab, je gr√∂√üer x wird.<br>
<br>
Der Logarithmus hat bei $x=1$ stets den Funktionswert 0, also $log_b(1)=0$, weil $b^0=1$.<br>
Er ist f√ºr negative Werte von $x$ nicht definiert.<br>
<br>
**Nat√ºrlicher Logarithmus zur Basis ùëí**<br>
Eine h√§ufig verwendete Logarithmusfunktion ist der nat√ºrliche Logarithmus, der die Basis ùëí (Euler‚Äôsche Zahl) verwendet und oft als $ln(x)$ geschrieben wird:<br>
$f(x)=ln(x)$<br>
Der nat√ºrliche Logarithmus beschreibt Prozesse, die in Relation zu ihrem aktuellen Wert wachsen oder abnehmen, wie z.B. das exponentielle Wachstum oder den radioaktiven Zerfall.<br>
<br>

Die allgemeine Logarithmusfunktion zur Basis $b$, $log_b(x)$, berechnet, wie oft man die Basis ùëè mit sich selbst multiplizieren muss, um ùë• zu erhalten. Sie ist definiert als:<br>
$log_b(x)= \frac{log(x)}{log(b)}$<br>
Da die Basis des Logarithmus f√ºr diese Berechnung im Z√§hler und Nenner dieselbe sein muss (z. B. Basis 10 oder die nat√ºrliche Basis ùëí), ist dieser Ausdruck universell anwendbar.<br>
Dieser Ausdruck ist besonders praktisch, da viele Programmiersprachen und Taschenrechner nur den nat√ºrlichen Logarithmus und den Zehner-Logarithmus (Basis 10) unterst√ºtzen.

```python
f_log = lambda x, b: np.log(x) / np.log(b)

x = np.linspace(1, 1000, 1000)
y1 = f_log(x, 2)
y2 = np.log10(x)
y3 = f_log(x, 5)
y4 = f_log(x,.5)

fig, ax = plt.subplots(ncols=4,figsize=(12,3))

ax[0].set_title("Nat√ºrlicher Log.")
ax[0].plot(x,y1, label="$f(x) = ln(x)$")
ax[0].legend()

ax[2].set_title("Dekadischer Log.")
ax[2].plot(x,y2, label="$f(x) = log_{10}(x)$")
ax[2].legend()

ax[1].set_title("Log. zur Basis 5")
ax[1].plot(x,y3, label="$f(x) = log_5(x)$")
ax[1].legend()

ax[3].set_title("fallend")
ax[3].plot(x,y4, label="$f(x) = log_{0,5}(x)$")
ax[3].legend();
```
