### Alternative mit Axes Objekten
Tupelobjekt entpacken und in zwei Variablen speichern:

```python
# Erstelle eine Figur mit 2x2 Subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(6, 4))
```

Damit man das Figur-Objekt und Achsenobjekt individuell bearbeiten kann.  
Man kan z.B. mit `ax`-Objekt plotten und/oder `figur`-Objekte als Bild speichern.

```python

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(6, 4))

# subplot(2,2,1) mit schwarzem Hintergrund und rotem Text
axs[0, 0].set_facecolor('grey')
axs[0, 0].text(
    0.5, 0.5, 'subplot(2,2,1)', 
    ha='center', va='center', 
    fontsize=18, alpha=0.5, color='red'
)

# subplot(2,2,4) mit gr√ºnem Hintergrund und gelbem Text
axs[1, 1].set_facecolor('green')
axs[1, 1].text(
    0.5, 0.5, 'subplot(2,2,4)', 
    ha='center', va='center', 
    fontsize=18, alpha=1.0, color='yellow'
)

# Zweiten und dritten Subplot deaktivieren
axs[0, 1].axis('off')
axs[1, 0].axis('off')

# Entferne die Achsenticks f√ºr den letzen Subplot
axs[1, 1].set_xticks([])
axs[1, 1].set_yticks([])
axs[1,1].set_title('Letzter')

fig.suptitle('Titel', fontsize=18)
plt.show()
```

2D Array "flachkloppen" und mit einer Schleife durchlaufen:

```python
# Erstelle eine Figur mit 2x2 Subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(6, 4))

for i, ax in enumerate(axs.flat):
    if i % 2 == 0:
        bcolor = "black"
        fcolor="white"
    else:
        bcolor = "white"
        fcolor= "black"
    
    ax.set_facecolor(bcolor)
    ax.text(
    0.5, 0.5, f'{i+1}. subplot', 
    ha='center', va='center', 
    fontsize=18, alpha=0.5, color=fcolor
    )
    ax.set_xticks([])
    ax.set_yticks([])
```

Verschieden gro√üe Subplots in einer Grafik mit `GridSpec`<br>
Ein GridSpec-Objekt in Matplotlib ist eine flexible M√∂glichkeit, ein Gitterlayout f√ºr Subplots zu erstellen. Es erm√∂glicht dir, Subplots in einer definierten Anzahl von Zeilen und Spalten anzuordnen, w√§hrend du die Gr√∂√üe und Position einzelner Subplots innerhalb dieses Rasters anpassen kannst.

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# Erstelle ein Figure-Objekt
fig = plt.figure(figsize=(10, 6))

# Erstelle ein GridSpec-Objekt
gs = gridspec.GridSpec(3, 3)

# F√ºge einen gro√üen Subplot hinzu, der √ºber mehrere Zeilen und Spalten geht
ax1 = fig.add_subplot(gs[0, :])  # Dieser geht √ºber die gesamte obere Reihe
ax1.set_title('Gro√üer Subplot')
ax1.plot([1, 2, 3], [1, 4, 9])

# F√ºge einen kleineren Subplot hinzu
ax2 = fig.add_subplot(gs[1, 0])  # Dieser ist in der ersten Spalte der zweiten Reihe
ax2.set_title('Kleiner Subplot 1')
ax2.plot([1, 2, 3], [1, 2, 3])

# F√ºge einen weiteren kleinen Subplot hinzu
ax3 = fig.add_subplot(gs[1, 1])  # Dieser ist in der zweiten Spalte der zweiten Reihe
ax3.set_title('Kleiner Subplot 2')
ax3.plot([1, 2, 3], [2, 3, 4])

# F√ºge einen weiteren Subplot hinzu
ax4 = fig.add_subplot(gs[1, 2])  # Dieser ist in der dritten Spalte der zweiten Reihe
ax4.set_title('Kleiner Subplot 3')
ax4.plot([1, 2, 3], [3, 2, 1])

# F√ºge einen gro√üen Subplot hinzu
ax5 = fig.add_subplot(gs[2, :2])  # Dieser geht √ºber die erste und zweite Spalte der dritten Reihe
ax5.set_title('Gro√üer Subplot 2')
ax5.plot([1, 2, 3], [1, 0, -1])

# F√ºge einen weiteren Subplot hinzu
ax4 = fig.add_subplot(gs[2, 2])  # Dieser ist in der dritten Spalte der dritten Reihe
ax4.set_title('Kleiner Subplot 4')
ax4.plot([1, 2, 3], [1, 2, 3])

# Passe die Layouts an
plt.tight_layout()
plt.show()
```

### 1. Rasterbildformate
Rasterbilder speichern Bildinformationen als Gitter aus einzelnen Bildpunkten, sogenannten Pixeln. Jedes Pixel enth√§lt Informationen √ºber eine spezifische Farbe und Helligkeit. Rasterbildformate sind ideal f√ºr komplexe, detailreiche Bilder wie Fotografien und digital erstellte Grafiken. 
- **JPEG** (Joint Photographic Experts Group) ist ein weit verbreitetes Bildformat, das sich besonders f√ºr Fotografien eignet. Es bietet eine hohe Farbtreue und eine relativ kleine Dateigr√∂√üe, was es ideal f√ºr das Web und digitale Fotografie macht. JPEG unterst√ºtzt keine Transparenz und verliert bei der Kompression einige Bilddetails, was zu einer sichtbaren Verschlechterung der Qualit√§t f√ºhren kann, wenn das Bild zu stark komprimiert wird.
- **PNG** (Portable Network Graphics) ist ein rasterbasiertes Format, das urspr√ºnglich als Ersatz f√ºr das GIF-Format entwickelt wurde. - **TIFF** (Tagged Image File Format) ist ein flexibles und vielseitiges Bildformat, das h√§ufig in der Druckindustrie und f√ºr hochwertige Scans verwendet wird. Es unterst√ºtzt sowohl verlustbehaftete als auch verlustfreie Kompression und kann mehrere Ebenen und Seiten in einer Datei speichern. Ideal f√ºr Bilder, die Transparenz erfordern, sowie f√ºr Grafiken mit scharfen Kanten und Text. \
Verlustfreie Kompression, wodurch die Bildqualit√§t erhalten bleibt. 
- **BMP** (Bitmap) ist ein einfaches, unkomprimiertes Bildformat, das Pixelinformationen in einer klaren Struktur speichert.
