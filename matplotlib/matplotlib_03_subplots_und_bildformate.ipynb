## Mehrere Diagramme mit eigenen Achsen

Wenn wir mehrere Diagramme separat und trotzedem in einem Fenster zeichnen wollen.

`subplot(nrows, ncols, plot_nr)`

- nrows: Anzahl der Zeilen
- ncols: Anzahl der Spalten
- plot_nr: Bezeichnung für ein Plot (zwischen 1 und nrows\*ncols)

`subplot(2, 3, 4)` kann auch so `subplot(234)`

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(6,4)) # 6 x 4 Zoll Leinwand
plt.subplot(221, facecolor='grey') # alternativ .subplot(2,2,1)
plt.text(
    0.5, # x-Koordinaten hier anfangen zu zeigen
    0.5, # y-Koordinate hier anfangen zu zeigen
    'subplot(2,2,1)', # Textbeschreibung
    ha = 'center', # Horizontal alignment (alternativ horizontalalignment='center')
    va = 'center', # Vertical alignment (alternativ verticalalignment='center')
    fontsize = 18, # Schriftgröße
    alpha = 0.5, # Durchsichtigkeit (zw. 0.0 und 1.0)
    color = 'red' # Textfarbe)
)
plt.subplot(2,2,4, facecolor='g') # Farbe des Hintergrunds
plt.text(
    0.5,
    0.5,
    'subplot(2,2,4)', # Textbeschreibung
    ha = 'center', # Horizontal alignment (alternativ horizontalalignment='center')
    va = 'center', # Vertical alignment (alternativ verticalalignment='center')
    fontsize = 18, # Schriftgröße
    alpha = 1.0, # Durchsichtigkeit (zw. 0.0 und 1.0)
    color = 'yellow' # Textfarbe)
)

plt.show()
```

### Alternative mit Axes Objekten
Tupelobjekt entpacken und in zwei Variablen speichern:

```python
# Erstelle eine Figur mit 2x2 Subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(6, 4))
```

Damit man das Figur-Objekt und Achsenobjekt individuell bearbeiten kann.  
Man kan z.B. mit `ax`-Objekt plotten und/oder `figur`-Objekte als Bild speichern.

```python

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(6, 4))

# subplot(2,2,1) mit schwarzem Hintergrund und rotem Text
axs[0, 0].set_facecolor('grey')
axs[0, 0].text(
    0.5, 0.5, 'subplot(2,2,1)', 
    ha='center', va='center', 
    fontsize=18, alpha=0.5, color='red'
)

# subplot(2,2,4) mit grünem Hintergrund und gelbem Text
axs[1, 1].set_facecolor('green')
axs[1, 1].text(
    0.5, 0.5, 'subplot(2,2,4)', 
    ha='center', va='center', 
    fontsize=18, alpha=1.0, color='yellow'
)

# Zweiten und dritten Subplot deaktivieren
axs[0, 1].axis('off')
axs[1, 0].axis('off')

# Entferne die Achsenticks für den letzen Subplot
axs[1, 1].set_xticks([])
axs[1, 1].set_yticks([])
axs[1,1].set_title('Letzter')

fig.suptitle('Titel', fontsize=18)
plt.show()
```

Figur-Objekte können als Bilddatei (File) gespeichert werden:

2D Array "flachkloppen" und mit einer Schleife durchlaufen:

```python
# Erstelle eine Figur mit 2x2 Subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(6, 4))

for i, ax in enumerate(axs.flat):
    if i % 2 == 0:
        bcolor = "black"
        fcolor="white"
    else:
        bcolor = "white"
        fcolor= "black"
    
    ax.set_facecolor(bcolor)
    ax.text(
    0.5, 0.5, f'{i+1}. subplot', 
    ha='center', va='center', 
    fontsize=18, alpha=0.5, color=fcolor
    )
    ax.set_xticks([])
    ax.set_yticks([])
```

Verschieden große Subplots in einer Grafik mit `GridSpec`<br>
Ein GridSpec-Objekt in Matplotlib ist eine flexible Möglichkeit, ein Gitterlayout für Subplots zu erstellen. Es ermöglicht dir, Subplots in einer definierten Anzahl von Zeilen und Spalten anzuordnen, während du die Größe und Position einzelner Subplots innerhalb dieses Rasters anpassen kannst.

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# Erstelle ein Figure-Objekt
fig = plt.figure(figsize=(10, 6))

# Erstelle ein GridSpec-Objekt
gs = gridspec.GridSpec(3, 3)

# Füge einen großen Subplot hinzu, der über mehrere Zeilen und Spalten geht
ax1 = fig.add_subplot(gs[0, :])  # Dieser geht über die gesamte obere Reihe
ax1.set_title('Großer Subplot')
ax1.plot([1, 2, 3], [1, 4, 9])

# Füge einen kleineren Subplot hinzu
ax2 = fig.add_subplot(gs[1, 0])  # Dieser ist in der ersten Spalte der zweiten Reihe
ax2.set_title('Kleiner Subplot 1')
ax2.plot([1, 2, 3], [1, 2, 3])

# Füge einen weiteren kleinen Subplot hinzu
ax3 = fig.add_subplot(gs[1, 1])  # Dieser ist in der zweiten Spalte der zweiten Reihe
ax3.set_title('Kleiner Subplot 2')
ax3.plot([1, 2, 3], [2, 3, 4])

# Füge einen weiteren Subplot hinzu
ax4 = fig.add_subplot(gs[1, 2])  # Dieser ist in der dritten Spalte der zweiten Reihe
ax4.set_title('Kleiner Subplot 3')
ax4.plot([1, 2, 3], [3, 2, 1])

# Füge einen großen Subplot hinzu
ax5 = fig.add_subplot(gs[2, :2])  # Dieser geht über die erste und zweite Spalte der dritten Reihe
ax5.set_title('Großer Subplot 2')
ax5.plot([1, 2, 3], [1, 0, -1])

# Füge einen weiteren Subplot hinzu
ax4 = fig.add_subplot(gs[2, 2])  # Dieser ist in der dritten Spalte der dritten Reihe
ax4.set_title('Kleiner Subplot 4')
ax4.plot([1, 2, 3], [1, 2, 3])

# Passe die Layouts an
plt.tight_layout()
plt.show()
```

## Bildformate
### 1. Rasterbildformate
Rasterbilder speichern Bildinformationen als Gitter aus einzelnen Bildpunkten, sogenannten Pixeln. Jedes Pixel enthält Informationen über eine spezifische Farbe und Helligkeit. Die Gesamtauflösung eines Rasterbildes ist dabei durch die Anzahl der Pixel festgelegt. Beim Vergrößern oder Verkleinern eines Rasterbildes bleibt die Pixelanzahl gleich, was zu Qualitätsverlust bei Vergrößerung führen kann. Rasterbildformate sind ideal für komplexe, detailreiche Bilder wie Fotografien und digital erstellte Grafiken. 
- **JPEG** (Joint Photographic Experts Group) ist ein weit verbreitetes Bildformat, das sich besonders für Fotografien eignet. Es bietet eine hohe Farbtreue und eine relativ kleine Dateigröße, was es ideal für das Web und digitale Fotografie macht. JPEG unterstützt keine Transparenz und verliert bei der Kompression einige Bilddetails, was zu einer sichtbaren Verschlechterung der Qualität führen kann, wenn das Bild zu stark komprimiert wird.
- **PNG** (Portable Network Graphics) ist ein rasterbasiertes Format, das ursprünglich als Ersatz für das GIF-Format entwickelt wurde. Es unterstützt eine verlustfreie Kompression und ermöglicht Transparenz, was es besonders nützlich für Grafiken, Logos und Bilder mit klaren Kanten macht. PNG behält die Bildqualität bei, auch wenn das Bild mehrmals gespeichert wird, und eignet sich gut für Webgrafiken.
- **TIFF** (Tagged Image File Format) ist ein flexibles und vielseitiges Bildformat, das häufig in der Druckindustrie und für hochwertige Scans verwendet wird. Es unterstützt sowohl verlustbehaftete als auch verlustfreie Kompression und kann mehrere Ebenen und Seiten in einer Datei speichern. TIFF eignet sich hervorragend für die Bearbeitung und Archivierung von Bildern, da es eine hohe Bildqualität bietet.
Ideal für Bilder, die Transparenz erfordern, sowie für Grafiken mit scharfen Kanten und Text. \
Verlustfreie Kompression, wodurch die Bildqualität erhalten bleibt. 
- **BMP** (Bitmap) ist ein einfaches, unkomprimiertes Bildformat, das Pixelinformationen in einer klaren Struktur speichert. BMP-Dateien sind in der Regel groß und bieten keine Unterstützung für Transparenz oder moderne Bildkompressionstechniken. Das Format wird vor allem in Windows-Umgebungen verwendet, ist jedoch aufgrund seiner großen Dateigröße weniger populär im Web oder in der digitalen Fotografie

**Unkomprimierte Rohbildformate (RAW)**:<br> Diese Formate speichern die unbearbeiteten Sensordaten einer Kamera und enthalten alle Bilddetails, was maximale Flexibilität bei der Nachbearbeitung erlaubt. Sie erfordern oft spezielle Software und beanspruchen viel Speicherplatz. Die Formate sind variabel und können je nach Kamerahersteller und Modell unterschiedlich sein.
<br><br>
**Animierte Bildformate**:<br> Diese Formate bestehen aus einer Sequenz von Rasterbildern, die in schneller Folge abgespielt werden, um eine Animation zu erzeugen. Animierte Formate sind beliebt für einfache Animationen,  Social-Media-Inhalte und Memes.
- **GIF** (Graphics Interchange Format) ist eines der bekanntesten animierten Formate. Es unterstützt bis zu 256 Farben pro Frame und ermöglicht einfache Animationen, indem mehrere Bilder in einer einzigen Datei gespeichert werden. GIFs sind beliebt für kurze, loopende Animationen, die häufig in sozialen Medien und auf Webseiten verwendet werden.
- **APNG** (Animated Portable Network Graphics) ist eine erweiterte Version des PNG-Formats, die Animationen unterstützt. Im Gegensatz zu GIF kann APNG eine größere Farbtiefe und Transparenz bieten. Es ist besonders nützlich für qualitativ hochwertige Animationen und wird von einigen modernen Browsern unterstützt.

**High-Efficiency-Formate**:<br> Diese Formate nutzen fortschrittliche Algorithmen zur Bildkomprimierung, um hohe Bildqualität bei minimaler Dateigröße zu ermöglichen. Sie sind oft auf moderne Anforderungen optimiert und unterstützen Funktionen wie Transparenz, Animation und HDR-Bilder (High Dynamic Range, Unterschied zwischen dem dunkelsten und dem hellsten Punkt). Diese Formate sind besonders für mobile Geräte und Web-Anwendungen geeignet, bei denen Speicher- und Übertragungseffizienz wichtig sind. 
- **HEIF** (High Efficiency Image File Format) ist ein Bildformat, das eine hohe Bildqualität bei geringerer Dateigröße als JPEG bietet. HEIF unterstützt sowohl verlustbehaftete als auch verlustfreie Kompression und ermöglicht die Speicherung von Bildern mit höherer Farbtiefe und mehr Bilddaten. Es wird auf Apple-Geräten verwendet (iPhones) und kann sowohl Einzelbilder als auch Animationen (ähnlich wie GIF) speichern.
- **WebP** ist ein Bildformat von Google, das sowohl verlustbehaftete als auch verlustfreie Kompression unterstützt. WebP kann die Dateigröße im Vergleich zu JPEG oder PNG erheblich reduzieren und unterstützt Transparenz. Es wird häufig für Webgrafiken verwendet, um die Ladezeiten zu optimieren.
- **FLIF** (Free Lossless Image Format) ist ein Bildformat, das verlustfreie Kompression bietet und dabei eine höhere Effizienz als andere Formate wie PNG oder TIFF aufweisen kann. FLIF ist darauf ausgelegt, eine hohe Bildqualität bei gleichzeitig geringer Dateigröße zu ermöglichen.

### 2. Vektorbildformate
Vektorbildformate speichern Bilddaten als mathematische Anweisungen, die Formen wie Linien, Kurven, Polygone und Farben definieren. Anstatt ein festes Raster zu haben, basiert ein Vektorbild auf geometrischen Berechnungen, die eine beliebige Skalierung ohne Qualitätsverlust ermöglichen. Diese Formate eignen sich besonders für Logos, Illustrationen, technische Zeichnungen und Icons, die in unterschiedlichen Größen und Auflösungen verwendet werden.<br>
- **SVG (Scalable Vector Graphics)**:
    Ein XML-basiertes Format, das für die Darstellung von Vektorgrafiken entwickelt wurde. \
    Da es auf Vektoren basiert, gibt es keine verlustbehaftete Komprimierung. \
    Ideal für Logos, Icons und Grafiken, die in verschiedenen Größen ohne Qualitätsverlust skaliert werden müssen. \

- **AI (Adobe Illustrator)**:
Ein proprietäres Vektorformat von Adobe, das in Adobe Illustrator verwendet wird.
Für professionelle Illustrationen und Designprojekte.
- **EPS (Encapsulated PostScript)**:
Ein Vektorformat, das häufig für den Druck und die Erstellung von Grafiken verwendet wird.
Verwendung: Ideal für den Austausch von Grafiken zwischen verschiedenen Anwendungen.
