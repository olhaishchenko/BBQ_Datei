`plot()` nimmt automatisch die Indizes von Elementen als Werte für die x-Achse und die Elemente selbst für die Werte entlang y-Achse.

`plot()` kann auch Datenpunkte (Data Points) wie Punkte darstellen.  
Dazu kann der Linienstil (linestyle) angepasst werden.

```python
ls = [-1, -3, -4.5, 5, 3, 4, 9]
plt.plot(ls, 'or') # o: Linienstil r: red
plt.show()
```

`plot()` akzeptiert auch ein weiteres Array für x-Achse:

```python
x = np.array([1,3,5,7,9])
y = np.array([2,6,10,14,18])
plt.plot(x,y,'bo')
plt.show()
```

Man kann auf derselben Leinwand mehrere Diagramme zeichnen:

```python
x = np.array([1,3,5,7,9])
y = np.array([2,6,10,14,18])
plt.plot(x,y) # Gerade
plt.plot(x,y,'or') # Rote Pünktchen
plt.show()
```

#### Bezeichnungen für die Achsen

- `plt.xlabel()`
- `plt.ylabel()`

```python
x = np.array([1,3,5,7,9])
y = np.array([2,6,10,14,18])
plt.plot(x,y) # Gerade
plt.plot(x,y,'or') # Rote Pünktchen
plt.xlabel('x-Achse') # Bezeichnung für die x-Achse
plt.ylabel('y-Achse') # Bezeichnung für die y-Achse
plt.show()
```

#### Rand anpassen

- `plt.margins()` Der Abstand zwischen Rand und Grafik (bei default 0.05)

```python
# Koordinate als Listen
x = [1,2,3]
y = [2,4,1]

# Rand anpassen (0,0)
# rechts und links: 0
# oben und unten:0 
plt.margins(0.0,0.0)

# plott erstellen
plt.plot(x,y)

# Achsenbezeichnungen und Überschrift
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('title')

# Grafik anzeigen
plt.show()
```

`ptl.legend()` \
MIt der Angabe von `label=` kann bei `plot()`ein Name angegeben werden der dann in der Legende erscheint. \
In der Datenvisualisierung geht es meistens darum, dass wir unseren Daten visuelle Darstellungen erstellen, um verschiedene Features zu einander zu vergleichen. Dazu zeichnen wir mehrere Grafiken auf einer Leinwand:

```python
x1 = [10,20,30]
y1 = [20,40,10]
plt.plot(x1, y1, 'go--', linewidth=2, markersize=10, label='1. Line')

x2 = [10,20,30]
y2 = [40,10,40]
plt.plot(x2, y2, 'ro-', linewidth=3, markersize=10, label='Line 2')

plt.legend()

plt.show()
```

#### Annotation

`plt.annotate()` schreibt einen Text neben Datenpunkten/Diagrammen<br>
1.Argument ist der Text der dargestellt werden soll, das 2.Argument ist die x,y Koordinate an dem der Text plaziert werden soll.<br>
Wenn der Text mit `xytext =` noch verschoben werden soll muss `textcoords =` angegeben werden.

```python
plt.plot([-3],[1],"o")
plt.annotate("Koordinate x=-3, y=1", (-3,1))
plt.show()
```

```python
# Koordinate erstellen
xs = np.arange(0,10,1)
ys = np.random.normal(loc=3, size=10, scale=0.4)
print('x-Werte:', xs)
print('y-Werte:', ys)
plt.plot(xs, ys, 'go')
plt.title('Kundenbewertungen für Produkt X')
plt.ylabel('Punkte')
plt.xlabel('Kunden(gruppen)')
for x,y in zip(xs,ys):
    label = "{:.2f}".format(y)
    plt.annotate(label, # der Text neben dem Datenpunkt
                 (x,y), # Koordinate (wo der Text stehen darf)
                 xytext = (20,0), # Abstand zwischen Text und Datenpunkt 
                 ha = 'center', # ha: horizontal alignment (horizontale Position von Text)
                 textcoords = "offset points" # relativ zu xytext
                )
plt.show()
```

X und Y Bereich einschränken:

```python
x = np.linspace(-10, 10, 1000)
y = x**3 + x + 25
plt.title("X Bereich von -10 bis 10, Y Bereich von -1000 bis 1000")
plt.plot(x,y)
plt.show()
```

Bei manaueller Anpassung werden keine Ränder mehr angezeigt

```python
plt.plot(x,y)
plt.title("X Bereich von -5 bis 5, Y Bereich von -450 bis 450")
plt.xlim([-5, 5])
plt.ylim([-450, 450])
plt.show()
```

X und Y "Ticks" (Werte die an den Strichen stehen) anpassen 

```python
plt.plot(x,y)
plt.title("X und Y Ticks anpassen")
plt.xlim([-5, 5])
plt.ylim([-450, 450])

plt.xticks(np.arange(-5, 6))             # alle ganzzahligen X Werte im Bereich anzeigen
plt.yticks([-350, -250, -50, 125, 375])  # beliebige Werte innerhalb des Y Bereichs anzeigen
plt.show()
```

### Balkendiagramm

- `plt.bar(koordinate)`  
Folgend ein Beispiel:

```python
x = np.array(list('ABCD'))
print(x)
y = np.array([10,20,30,25]) 
print(y)
```

##### Säulendiagramm
Wenn Balken zu x-Achse gerade stehen.

```python
# bar()
plt.bar(x,y,color='magenta', width=0.6)
plt.show()
```

##### Horizontale Balkendiagramme

```python
# barh(): balkendiagramm horizontal
plt.barh(x,y,color='skyblue', height=0.6)
plt.show()
```

#### Histogramm

zeigt die Verteilung eines Wertes entlang eines weiteren Features an.  
Die Entwicklung von Anzahl der Arbeitslosen in einer gewissen Zeitspanne.

- `plt.hist()`

##### Beispiel
In einer Studie haben wir 100 Personen aus verscheidenen Altersgruppen auf eine besondere gesundheitliche Kondition (z.B. Mangel an Vitamin D) untersucht. Diese Personen sind zwischen 25 und 85 Jahre alt.

```python
ages = np.random.randint(25, 85, 100)
```

```python
ages = np.array([75, 82, 47, 32, 53, 80, 55, 73, 51, 52, 28, 44, 38, 30, 25, 49, 26,
       66, 31, 79, 44, 67, 37, 79, 79, 58, 30, 62, 65, 53, 51, 63, 34, 73,
       63, 55, 73, 30, 30, 82, 74, 48, 82, 82, 39, 68, 77, 28, 29, 44, 84,
       80, 64, 37, 46, 69, 78, 81, 41, 48, 35, 29, 76, 63, 77, 48, 81, 29,
       35, 45, 71, 72, 66, 75, 58, 32, 73, 66, 30, 71, 75, 46, 75, 25, 64,
       84, 47, 82, 73, 25, 39, 71, 35, 51, 25, 43, 52, 77, 46, 26])
```

```python
num_bins = 5
plt.hist(ages, num_bins, facecolor='gray', alpha=0.5)
plt.xlabel('Altersgruppen')
plt.ylabel('Anzahl der Personen')
plt.title('Verteilung Patienten in versch. Altersgruppen.')
plt.show()
```

Anhand unseres Diagramm können wir behaupten, dass Menschen in der Altersgruppe zwischen 50 und 60 am wenigsten mit dieser gesundheitlichen Kondition zu tun haben, im Gegenteil müssen die Menschen über 70 sich wesentlich mehr damit auseinandersetzen.
