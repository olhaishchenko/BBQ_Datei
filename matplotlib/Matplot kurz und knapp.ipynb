**Matplotlib:**
 Kurze Übersicht

**Definition:**
Matplotlib ist eine weit verbreitete Python-Bibliothek zur Erstellung von 2D-Graphen und -Diagrammen.
Sie ermöglicht es Benutzern, Daten visuell darzustellen und bietet viele Optionen zur Anpassung von Grafiken.

**Anwendung:**

1. **Datenvisualisierung:** Diagramme, Plots und Grafiken zur Analyse von Daten.
2. **Wissenschaftliche Berechnungen:** Unterstützung von Forschung und Analyse durch visuelle Darstellung von Ergebnissen.


**Vorteile:**

1. **Flexibel und anpassbar:** Eine Vielzahl von Diagrammtypen (Linien-, Balken-, Streu- und Kreisdiagramme).
2. **Integration:** Lässt sich gut mit anderen Bibliotheken wie NumPy und Pandas kombinieren.
3. **Dokumentation:** Umfassende und leicht verständliche Dokumentation.


**Nachteile:**

1. **Komplexität:** Kann für Anfänger überwältigend sein, da viele Optionen zur Verfügung stehen.
2. **Performance:**  Bei sehr großen Datenmengen kann die Performance leiden.

**Beispiel**
Hier ist ein einfaches Beispiel, wie man Matplotlib verwendet, um ein Liniendiagramm zu erstellen:

```python
import matplotlib.pyplot as plt

# Beispiel-Daten
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Liniendiagramm erstellen
plt.plot(x, y)
plt.title('Einfaches Liniendiagramm')
plt.xlabel('X-Achse')
plt.ylabel('Y-Achse')
plt.grid(True)
plt.show()
```

**Häufig verwendete Optionen von Matplotlib:**

1. **Plottypen**

1.1 **Linienplot (plot):**
 Dient zur Darstellung von Datenpunkten in einer Linie.

Beispiel:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
plt.plot(x, y)  # Liniendiagramm
plt.title("Linienplot")
plt.show()
```

1.2. **Balkendiagramm (bar):**  Zeigt kategorische Daten mit Balken.

Beispiel:

```python
categories = ['A', 'B', 'C']
values = [3, 7, 5]
plt.bar(categories, values)  # Balkendiagramm
plt.title("Balkendiagramm")
plt.show()
```

1.3. **Kreisdiagramm (pie):** Visualisiert die Verteilung von Daten in Prozent.

Beispiel:

```python
labels = ['A', 'B', 'C']
sizes = [30, 50, 20]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')  # Kreisdiagramm
plt.title("Kreisdiagramm")
plt.show()
```

2. **Anpassungen**

2.1 **Achsenbeschriftungen:**
Beschriftungen für die X- und Y-Achsen hinzufügen.

Beispiel:

```python
plt.plot(x, y)
plt.xlabel('X-Achse')  # X-Achsenbeschriftung
plt.ylabel('Y-Achse')  # Y-Achsenbeschriftung
plt.title("Mit Achsenbeschriftungen")
plt.show()
```

2.2 **Legende:**

 Erklärt die Bedeutung der Linien oder Balken.

Beispiel:

```python
plt.plot(x, y, label='Quadratwerte')
plt.legend()  # Legende anzeigen
plt.title("Legende Beispiel")
plt.show()
```

2.3. **Rasterlinien:**

 Macht es einfacher, Datenpunkte zu lesen.

Beispiel:

```python
plt.plot(x, y)
plt.grid(True)  # Rasterlinien aktivieren
plt.title("Mit Rasterlinien")
plt.show()
```

3. **Farben und Stile**

**Farben anpassen:** Man kann die Farbe der Linien und Balken anpassen.

Beispiel:

```python
plt.plot(x, y, color='red', linestyle='--')  # Rote gestrichelte Linie
plt.title("Farb- und Stiländerung")
plt.show()
```

3.1. **Farben ändern**

Es gibt mehrere Möglichkeiten, Farben in Matplotlib anzupassen:

Direkte Farbnamen: Du kannst vordefinierte Farbnamen verwenden, z. B. 'red', 'blue', 'green'.

```python
plt.plot([1, 2, 3, 4], color='green')  # Grüne Linie
```

3.2. **Hexadezimale Farbwerte:**

 Verwende Farbcodes im Hex-Format, wie z. B. '#FF5733' für Orange.

```python
plt.plot([1, 2, 3, 4], color='#FF5733')  # Orange Linie
```

3.3. **RGB-Tupel:**

 Definiere Farben mit einem RGB-Tupel, z. B. (0.1, 0.2, 0.5).

```python
plt.plot([1, 2, 3, 4], color=(0.1, 0.2, 0.5))  # Dunkelblaue Linie
```

3.4. **Linienstile ändern**

Die Linie kann auf verschiedene Arten angepasst werden:

Linienstile: Du kannst verschiedene Linienstile wie durchgezogene, gestrichelte oder gepunktete Linien verwenden.

'solid' oder '-': Durchgezogene Linie (Standard).

'dashed' oder '--': Gestrichelte Linie.

'dotted' oder ':': Gepunktete Linie.

'-.': Gemischte Strich-Punkt-Linie.

```python
plt.plot([1, 2, 3, 4], linestyle='--')  # Gestrichelte Linie
plt.plot([1, 2, 3, 4], linestyle='-.')  # Strich-Punkt-Linie
```

3.5 **Linienbreite ändern**

Du kannst die Dicke der Linie mit dem Parameter linewidth ändern.

```python
plt.plot([1, 2, 3, 4], linewidth=2.5)  # Dickere Linie
```

3.6 **Marker hinzufügen**

Marker sind Symbole, die an den Datenpunkten entlang einer Linie angezeigt werden können. Hier sind einige Marker-Optionen:

'o': Kreis

'^': Dreieck

's': Quadrat

'x': Kreuz

```python
plt.plot([1, 2, 3, 4], marker='o', linestyle='--')  # Gestrichelte Linie mit Kreismarkern
```

3.7. **Mehrere Anpassungen kombiniert**

Hier ist ein Beispiel, das Farben, Linienstile, Linienbreiten und Marker kombiniert:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

# Kombinieren von Farbe, Linienstil, Marker und Linienbreite
plt.plot(x, y, color='purple', linestyle='-.', marker='o', linewidth=2)
plt.title("Angepasste Linie")
plt.show()
```

4. **Unterplots**
Mehrere Diagramme in einem: Erstelle ein Gitter von Diagrammen.

Beispiel:

```python
plt.subplot(1, 2, 1)  # 1 Zeile, 2 Spalten, 1. Diagramm
plt.plot(x, y)
plt.title("Diagramm 1")

plt.subplot(1, 2, 2)  # 1 Zeile, 2 Spalten, 2. Diagramm
plt.bar(categories, values)
plt.title("Diagramm 2")

plt.tight_layout()  # Abstand zwischen den Diagrammen optimieren
plt.show()
```

5. **Kartesisches Koordinatensystem.**

 Ein kartesisches Koordinatensystem beschreibt die Position von Punkten in einer Ebene (2D) oder im Raum (3D) durch geordnete Zahlenpaare oder -tripel. Es wurde von René Descartes eingeführt.

1. **2D-Koordinatensystem:**
Besteht aus zwei Achsen:

x-Achse (horizontal)

y-Achse (vertikal)

Der Ursprung ist der Punkt (0, 0), wo sich die Achsen schneiden.
Ein Punkt wird als (x, y) dargestellt, wobei x die horizontale und y die vertikale Position angibt.

2. **3D-Koordinatensystem:**

Hat drei Achsen:

x-Achse (horizontal)

y-Achse (vertikal)

z-Achse (Tiefenachse)

Ein Punkt wird durch (x, y, z) beschrieben, wobei z die Position in der Tiefe angibt.

```python
import matplotlib.pyplot as plt

# Zeichnen des kartesischen Koordinatensystems
plt.axhline(0, color='black')  # X-Achse
plt.axvline(0, color='black')  # Y-Achse

# Punkt (3, 4) darstellen
plt.scatter(3, 4, color='red')  # Punkt bei (3, 4)
plt.text(3, 4, "(3,4)", fontsize=12)  # Beschriftung des Punktes

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.grid(True)  # Gitterlinien anzeigen
plt.title("Kartesisches Koordinatensystem")
plt.show()
```
