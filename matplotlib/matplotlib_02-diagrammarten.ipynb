#### Histogramm

werden zur Darstellung von Häufigkeitsverteilungen eines Features (Merkmals) verwendet. Eine Videoerklärung vom Begriff 'Histogramm' findest du [hier](https://studyflix.de/statistik/histogramm-1711). Um ein Histogramm zu zeichnen, müssen Daten in verschiedenen Kategorien verteilt werden. Im folgenden Beispiel erstellen wir anhand Zufallzahlen ein Histogramm.  

Ein Histogramm kann mit Hilfe der Funktion `hist()` aus dem Modul `pyplot` erstellt werden:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
gaussian_numbers = np.random.normal(size=10000)
plt.hist(gaussian_numbers)
plt.title("Gaussian Histogram")
plt.xlabel("Wert")
plt.ylabel("Häufigkeit")
plt.show()
```

Dieses Histogramm veraunschaulicht eine [_Normalverteilung_](https://de.wikipedia.org/wiki/Normalverteilung) von 10000 Zufallzahlen. Dieser Art Verteilung heißt in der Stochastik auch Gaußsche Verteilung (nach [Carl Friedrich Gauß](https://de.wikipedia.org/wiki/Carl_Friedrich_Gau%C3%9F)).

#### Tortendiagramm
Ein Tortendiagramm, auch Kreisdiagramm genannt (engl. pie plot) ist eine kreisförmige Darstellung (etwa prozentualer Verteilungen) für Teilwerte eines Ganzen als Teile eines Kreises. Das Kreisdiagramm ist kreisförmig und in mehrere Kreissektoren eingeteilt, wobei jeder Kreissektor einen Teilwert und der Kreis somit die Summe der Teilwerte (das Ganze) darstellt. Eine Videoerklärung dazu findest du [hier](https://studyflix.de/statistik/kreisdiagramm-1873)  

Kreisdiagramme erstellt man mit Hilfe der Funktion `pie()` aus dem Modul `pyplot`.

Im folgenden Beispiel erstellen wir ein Tortendiagramm:

```python
labels = ['SPD', 'CDU', 'Linke', 'Grüne', 'AFD', 'FDP', 'Sonstiges']
sizes = [38, 31, 27, 27, 22, 12, 3]
colors = ['red', 'black', 'darkred', 'green', 'cornflowerblue', 'yellow', 'gray'] # matplotlib named colors: https://matplotlib.org/stable/gallery/color/named_colors.html

explode = (0.2, 0.1, 0, 0, 0, 0, 0)  

plt.figure(figsize=(8,8)) # Die Größe des Diagramms bestimmen (Zoll - inch)

wedges, texts, autotexts = plt.pie(                             # wir brauchen nur autotexts um die Schrift anzupassen
        sizes,  # Größe der Tortenteile individuell bestimmen
        explode=explode, 
        labels=labels,
        shadow=True,
        startangle=90, # Beschriftung (Label) für jeden Kuchenteil
        autopct='%1.1f%%',
        colors = colors) # autopct: berechnet automatisch den Prozentwert jedes Teils aus dem Ganzen

for i, label in enumerate(labels):
    if label == 'CDU':
        autotexts[i].set_color('white') 

plt.title('Sitzverteilung im Berliner Abgeordnetenhaus 2016')
plt.show()
```

```python
sizes = [38, 31, 27, 27, 22, 12, 3]
print('Die Anzahl der Sitze:',sum(sizes))
```

#### Box-Plot

Der Box-Plot (auch Box-Whisker-Plot oder deutsch Kastengrafik) ist ein Diagramm, das zur grafischen Darstellung der Verteilung eines Merkmals verwendet wird. Eine Videoerklärung dazu findest du [hier]()  
Im folgenden Beispiel erstellen wir ein Box-Plot mit Hilfe der Funktion `box` aus dem Modul `pyplot`

```python
x1 = np.random.normal(0, 100, 1000)
plt.boxplot(x1)
plt.show()
```

```python
plt.hist(x1)
plt.show()
```

#### Streudiagramme
Ein Streudiagramm, auch Punktwolke genannt (engl. scatter plot), ist die graphische Darstellung von beobachteten Wertepaaren zweier statistischer Merkmale. Diese Wertepaare werden in ein kartesisches Koordinatensystem eingetragen, wodurch sich eine Punktwolke ergibt. Die Darstellung der Punkte kann durch verschiedene Symbole erfolgen.

```python
x = [5,3,7,8,7,2,17,2,9,4,11,12,9,6] # Alterzahlen von gewissen Autos
y = [99,70,86,87,88,111,86,103,87,94,78,77,85,86] # die Max. Geschwindigkeiten
plt.xlabel('Alter')
plt.ylabel('Geschwindigkeit m/h')
plt.title('Je älter, desto langsamer ... ')
plt.scatter(x,y, alpha=0.5)
plt.show()
```
