## Einfache Datenvisualisierung 

#### Übung
Für die ersten 22 Tage eines Monats haben wir den Durchschnitt-Lufttemperaturen in Grad Celsius für alle jeweils drei Tage gemessen und in einer Liste gespeichert:

```python
celsius_values = [25.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
```
Erstelle dafür mit Matplotlib ein Diagramm, welches entlang x-Achse die Tage (z.B. 1 bis 22) und entlang der y-Achse die Temperaturen angezeigt werden.

````python
days = list(range(1,23,3))
len(days)
````

````python
celsius_values = [25.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
len(celsius_values)
````

````python
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
plt.plot(days, celsius_values, 'go')
plt.xticks(days)
plt.title("Temperaturen als Streudiagramm")
plt.xlabel("Tage")
plt.ylabel("Durchschnittstemperatur")
plt.show()
````

````python
plt.plot(days, celsius_values)
plt.xticks(days)
plt.title("Temperaturen als Liniendiagramm")
plt.xlabel("Tage")
plt.ylabel("Durchschnittstemperatur")
plt.show()
````

````python
plt.plot(days, celsius_values, 'go')
plt.plot(days, celsius_values)
plt.xticks(days)
plt.title("Temperaturen als Streu- und Liniendiagramm")
plt.xlabel("Tage")
plt.ylabel("Durchschnittstemperatur")
plt.show()
````

````python
plt.bar(days, celsius_values)
plt.title("Temperaturen als Balkendiagramm/Säulendiagramm")
plt.xlabel("Tage")
plt.xticks(days)
plt.ylabel("Durchschnittstemperatur")
plt.show()
````
