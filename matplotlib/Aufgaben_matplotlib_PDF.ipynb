```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0,50)
y = x*3

plt.figure(figsize=(10,8))
plt.plot(x,y)
plt.title("Draw a line")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.tick_params(axis='both', top='both', right='both', direction='in');
```

```python
# Koordinate als Listen
x = [1,2,3]
y = [2,4,1]

# Rand anpassen (0,0)
# rechts und links: 0
# oben und unten:0 
plt.margins(0.0,0.0)

# plott erstellen
plt.plot(x,y)

# Achsenbezeichnungen und berschrift
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Sample graph!')
plt.xticks(np.arange(1,3.5,0.5)) 
# Grafik anzeigen
plt.show()
```

```python
plt.margins(0.0,0.0)
x1 = [10,20,30]
y1 = [20,40,10]
plt.plot(x1, y1, 'g', linewidth=1, label='line 1')

x2 = [10,20,30]
y2 = [40,10,30]
plt.plot(x2, y2, 'b', linewidth=1, label='line 2')

plt.xticks(np.arange(10,35,5)) 
plt.legend(loc="upper right")

plt.title('Two or more lines on same plot with suitable legends')
plt.show()
```

```python
plt.margins(0.0,0.0)
x1 = [10,20,30]
y1 = [20,40,10]
plt.plot(x1, y1, 'b', linewidth=3, label='line 1-width-3')

x2 = [10,20,30]
y2 = [40,10,30]
plt.plot(x2, y2, 'r', linewidth=5, label='line 2-width-5')

plt.xticks(np.arange(10,35,5)) 
plt.legend(loc="upper right")

plt.title('Two or more lines with different widths and colors with suitable legends')
plt.show()
```

```python
plt.margins(0.0,0.0)
x1 = [10,20,30]
y1 = [20,40,10]
plt.plot(x1, y1, 'b', ls='dotted', linewidth=3, label='line 1-dotted')

x2 = [10,20,30]
y2 = [40,10,30]
plt.plot(x2, y2, 'r', ls='dashed', linewidth=4, label='line 2-dashed')

plt.xticks(np.arange(10,35,5)) 
plt.legend(loc="upper right")

plt.title('Plot with two or more lines with different styles')
plt.show()
```

```python
plt.margins(0.0,0.0)
x1 = [1,4,5,6,7]
y1 = [2,6,3,6,3]
plt.plot(x1, y1, 'r', ls=':', linewidth=2, marker='o', markersize=8, markerfacecolor="blue")

plt.xticks(np.arange(1,9)) 
plt.yticks(np.arange(1,9)) 

plt.title('Display marker')
plt.show()
```

```python

x1 = [3,4,6,7,9]
y1 = [2,6,11,20,22]
plt.scatter(x1, y1, alpha=0.7, color='r', marker='o')
# plt.scatter(x1, y1, alpha=0.5, color='b', marker='x', label="Datenpunkte")

x2 = [2,3,5,6,8]
y2 = [1,5,10,18,20]
plt.scatter(x2, y2, alpha=0.4, color='b', marker='*')

plt.xticks(np.arange(0,11,2)) 
plt.yticks(np.arange(0,35,5)) 

plt.show()
```

```python
plt.figure(figsize=(10,8))
x = np.arange(0., 5., 0.2)
y1 = x**2
y2 = x**3
y3 = x


plt.plot(x, y1, 'bs', label='y = x^2')  
plt.plot(x, y2, 'r^', label='y = x^3') 
plt.plot(x, y3, 'g--', label='y = x')   


plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc="upper left")


plt.show()
```

```python
import matplotlib.gridspec as gridspec
```

```python
# Erstelle ein Figure-Objekt
fig = plt.figure(figsize=(8, 6))

# Erstelle ein GridSpec-Objekt
gs = gridspec.GridSpec(2, 3)

# Fge einen groen Subplot hinzu, der ber mehrere Zeilen und Spalten geht
ax1 = fig.add_subplot(gs[0, :])  # Dieser geht ber die gesamte obere Reihe
ax1.plot()
ax1.set_xticks(())
ax1.set_yticks(())

# Fge weiteren kleinen Subplots hinzu
for i in range(3):
    axs = fig.add_subplot(gs[1, i]) # Dieser ist in der zweiten Spalte der zweiten Reihe
    axs.plot()
    axs.set_xticks([])
    axs.set_yticks([])

# Passe die Layouts an
# plt.tight_layout()
plt.show()
```
