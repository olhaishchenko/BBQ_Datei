(b) Erstellen Sie mit Matplotlib ein Balkendiagramm (oder Tortendiagramm), um die Popularität der o.g. Sprachen grafisch zu visualisieren.

````python
# code
plt.figure(figsize=(14,10))
fig, ax = plt.subplots()

hbars = ax.barh(programmiersprachen, ratings, align='center')
ax.set_yticks(programmiersprachen, labels=programmiersprachen)
ax.bar_label(hbars, labels=[f'{r:.2f}%' for r in ratings],
             padding=8, color='r', fontsize=10)
ax.set_title('Ranking von Programmiersprachen')
ax.set_ylim(top = len(programmiersprachen))

ax.set_xlim(right=25) ;
````

(a) Schreiben Sie ein Programm mit NumPy Arrays, das 100 Würfe mit zwei Würfeln simuliert.   
(b) Die Verteilung der gewürfelten Zahlen soll alsHistogramm dargestellt werden.
<br>
<br>
<img src='histo.png' width='350px' align="center"/>
<br>
Ein Histogramm, das die Verteilung von Würfen mit zwei Würfeln anzeigt.


````python
# code
import numpy as np
import matplotlib.pyplot as plt

# (a) Simuliere 100 Würfe mit zwei Würfeln
np.random.seed(0)  # Für Reproduzierbarkeit
dice_rolls = np.random.randint(1, 7, (100, 2))  # Zwei Würfel, 100 Würfe
sums = dice_rolls.sum(axis=1)  # Summe der beiden Würfel für jeden Wurf

# (b) Erstelle das Histogramm
plt.figure(figsize=(8, 6))
plt.hist(sums, bins=range(2, 14), edgecolor='black', align='left', color='skyblue')

# Achsentitel und Diagrammtitel hinzufügen
plt.xlabel("Summe der beiden Würfel")
plt.ylabel("Häufigkeit")
plt.title("Verteilung der Summe bei 100 Würfen mit zwei Würfeln")
plt.xticks(range(2, 13))

# Diagramm anzeigen
plt.show()
````
