Explanation
- FacetGrid is used to create separate plots (facets) for each value of the Sex column.
- sns.histplot is mapped to each facet to plot the age distribution.
- kde=True adds a Kernel Density Estimate (KDE) line to visualize the distribution smoothly.
- stat="density" normalizes the histogram to show density rather than count, similar to the plot in your image.

```python
# Hier eigenen Code schreiben
# Nächste Zelle nicht nutzen
g = sns.FacetGrid(titanic, col="sex")
g.map(sns.histplot, "age", kde=True, stat="density")
# g = g.map(plt.hist, "age")
```
